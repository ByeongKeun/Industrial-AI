```
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```
iris = load_iris()
descision_tree = DecisionTreeClassifier(criterion="entropy", random_state=0, max_depth=3)
descision_tree = descision_tree.fit(iris.data, iris.target)
```

```
plt.figure()
plot_tree(descision_tree, filled=True)
plt.show()
```
