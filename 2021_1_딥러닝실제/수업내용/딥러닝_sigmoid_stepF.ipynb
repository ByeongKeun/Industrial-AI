```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
def step_function(x):
  return np.array(x>0, dtype=np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x)
y2 = step_function(x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2, 'k--')
plt.ylim(-0.1, 1.1) #y축 범위 지정
plt.show()
```

```python
def relu(x):
  return np.maximum(0, x)
```

```python
#relu 함수
x4 = np.arange(-5, 5, 0.1)
y4 = relu(x4)
plt.plot(x4, y4, 'b--')
plt.show()
```
